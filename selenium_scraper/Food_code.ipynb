```python
def convert_to_number(input_str):
    suffixes = {
        'T': 10**12,
        'B': 10**9,
        'M': 10**6
    }

    # Extract the numeric value and suffix from the input string
    value_str, suffix = input_str[:-1], input_str[-1].upper()

    # Multiply the numeric value by the corresponding suffix
    if suffix in suffixes:
        result = float(value_str) * suffixes[suffix]
        return result
    else:
        print("Invalid suffix. Use T, B, or M.")
        return None

# Apply the conversion function to the "Revenue" column
df['Revenue'] = df['Revenue'].apply(convert_to_number)

pd.options.display.float_format = '{:.2f}'.format

# Display the updated DataFrame
print(df)
```
