```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
rank_list=[]
category=[]
company=[]
country=[]
market_capital=[]
```

##  Market Cap

```python

# Set up the Chrome WebDriver (you need to have ChromeDriver installed and in your PATH)




def loop():
    global c_i
    global company
    global country
    
    
    company.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[3]/div[2]/a/div[1]').text)

    market_capital.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[4]').text)

    country.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[8]/span').text)
       
        



# page iteration
for pg_id in range(1,2): 
    driver = webdriver.Chrome()

    url = f"https://companiesmarketcap.com/e-commerce/largest-e-commerce-companies-by-market-cap/"
    driver.get(url)        

    c_i = 1

    for i in range(1,96):

        r_iter=i
        
        cat='üõí E-Commerce'
        category.append(cat)
        
        if pg_id == 1:
            rank=i
            
        elif pg_id == 2:
            rank=i+100
           
        elif pg_id == 3:
            rank=i+200
            
        elif pg_id == 4:
            rank=i+300   
            
        elif pg_id == 5:
            rank=i+400 
        
        elif pg_id == 6:
            rank=i+500 
         
        else:
            rank=i+600 
            
        rank_list.append(rank)
            

        if r_iter==18 or r_iter == 76:
            c_i +=1
            loop()

        else:
            loop()

        c_i+=1

    driver.close()     
            


```

```python
data = {'Rank':rank_list,'Category': category, 'Company': company, 'Market_Capital': market_capital,'Country':country}
df_MP=pd.DataFrame(data)
df_MP
```

```python
df_MP = df_MP [df_MP.Market_Capital != 'N/A']
```

```python
df_MP['Market_Capital']=df_MP['Market_Capital'].str.replace('$','')
```

```python
df_MP.to_csv('Ecommerce_MarketCapital.csv',index=False)
```
