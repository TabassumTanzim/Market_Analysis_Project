```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
# Set up the Chrome WebDriver (you need to have ChromeDriver installed and in your PATH)


def loop():
    global c_i
 
    
    food_country.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[8]/span').text)

    food_earning.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[4]').text)

        



# page iteration
for pg_id in range(3,4): 
    driver = webdriver.Chrome()

    url = f"https://companiesmarketcap.com/food/most-profitable-food-companies/?page={pg_id}"

    driver.get(url)        

    c_i = 1

    for i in range(1,40):

        r_iter=i
       
        cat='üç¥ Food'
        food_category.append(cat)
        
        
        if pg_id == 1:
            rank=i
            
        elif pg_id == 2:
            rank=i+100
           
        elif pg_id == 3:
            rank=i+200
            
        elif pg_id == 4:
            rank=i+300   
            
        elif pg_id == 5:
            rank=i+400 
        
        elif pg_id == 6:
            rank=i+500 
         
        else:
            rank=i+600 
            

        if r_iter==18 or r_iter == 76:
            c_i +=1
            loop()

        else:
            loop()

        c_i+=1

    driver.close()     
            

```

```python
data = {'Category': food_category, 'Earning': food_earning,'Country': food_country}
food_rev=pd.DataFrame(data)
food_rev=food_rev[food_rev.Earning !='N/A']


food_rev
```
