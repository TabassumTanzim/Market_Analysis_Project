## Market Cap

```python
import pandas as pd
```

```python
pharma=pd.read_csv('Pharma_MarketCapital2.csv')
```

```python
pharma['Market Capital']=pharma['Market Capital'].str.replace('$','')
```

```python
pharma=pharma.rename(columns={'Market Capital':'Market_Capital'})
```

```python
pharma
```

```python
food=pd.read_csv('Food_MarketCapital.csv')
food
```

```python
auto=pd.read_csv('Automobile_MarketCapital.csv')
auto
```

```python
auto
```

```python
ecom=pd.read_csv('Ecommerce_MarketCapital.csv')
ecom
```

```python

res=pd.read_csv('Res_MarketCapital.csv')
res
```

```python
MarketCap =pd.concat([pharma,food,auto,ecom,res],ignore_index=True)
```

```python
def convert_to_number(input_str):
    suffixes = {
        'T': 10**12,
        'B': 10**9,
        'M': 10**6
    }

    # Extract the numeric value and suffix from the input string
    value_str, suffix = input_str[:-1], input_str[-1].upper()

    # Multiply the numeric value by the corresponding suffix
    if suffix in suffixes:
        result = float(value_str) * suffixes[suffix]
        return result
    else:
        print("Invalid suffix. Use T, B, or M.")
        return None

# Apply the conversion function to the "Revenue" column
MarketCap['Market_Capital'] = MarketCap['Market_Capital'].apply(convert_to_number)

pd.options.display.float_format = '{:.2f}'.format

# Display the updated DataFrame
print(MarketCap)
```

```python
MarketCap.to_csv('MarketCap.csv',index=False)
```
