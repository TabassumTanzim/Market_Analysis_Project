### Food

```python
food_category=[]
food_country=[]
food_revenue=[]
```

```python

# Set up the Chrome WebDriver (you need to have ChromeDriver installed and in your PATH)
#FOOD

from selenium import webdriver
from selenium.webdriver.common.by import By



def loop():
    global c_i
    
    
#     company.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[3]/div[2]/a/div[1]').text)

    food_revenue.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[4]').text)

    food_country.append(driver.find_element(By.XPATH,'//*[@id="cmkt"]/div[5]/table/tbody/tr[' + str(c_i) + ']/td[8]/span').text)


        



# page iteration
for pg_id in range(3,4): 
    driver = webdriver.Chrome()

    url = f"https://companiesmarketcap.com/food/largest-food-companies-by-revenue/?page={pg_id}"
    driver.get(url)        

    c_i = 1

    for i in range(1,40):

        r_iter=i
        
        cat=' üç¥ Food'
        food_category.append(cat)
        
        if pg_id == 1:
            rank=i
            
        elif pg_id == 2:
            rank=i+100
           
        elif pg_id == 3:
            rank=i+200
        
        elif pg_id == 4:
            rank=i+300   
            
        elif pg_id == 5:
            rank=i+400 
        
        elif pg_id == 6:
            rank=i+500 
         
        else:
            rank=i+600 
            
            

        if r_iter==18 or r_iter == 76:
            c_i +=1
            loop()

        else:
            loop()

        c_i+=1

    driver.close()     
            




```

```python
data = {'Category': food_category, 'Revenue': food_revenue,'Country': food_country}
df_rev=pd.DataFrame(data)
df_rev
```
